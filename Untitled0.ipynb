```python
# !echo "import torch" | cat - /content/HackByte_Dataset/train.py > temp && mv temp /content/HackByte_Dataset/train.py

train_script = """
import argparse
import os
os.environ["MPLBACKEND"] = "Agg"
import torch
from ultralytics import YOLO

# Default hyperparameters
EPOCHS = 50
MOSAIC = 0.1
OPTIMIZER = 'AdamW'
MOMENTUM = 0.2
LR0 = 0.001
LRF = 0.0001
SINGLE_CLS = False

if __name__ == '__main__':
    parser = argparse.ArgumentParser()
    parser.add_argument('--epochs', type=int, default=EPOCHS)
    parser.add_argument('--mosaic', type=float, default=MOSAIC)
    parser.add_argument('--optimizer', type=str, default=OPTIMIZER)
    parser.add_argument('--momentum', type=float, default=MOMENTUM)
    parser.add_argument('--lr0', type=float, default=LR0)
    parser.add_argument('--lrf', type=float, default=LRF)
    parser.add_argument('--single_cls', type=bool, default=SINGLE_CLS)
    args = parser.parse_args()

    this_dir = os.path.dirname(__file__)
    os.chdir(this_dir)

    # Automatically select CPU or GPU
    device = 'cuda' if torch.cuda.is_available() else 'cpu'

    # Load YOLO model
    model = YOLO(os.path.join(this_dir, "yolov8s.pt"))

    # Train the model
    results = model.train(
        data=os.path.join(this_dir, "yolo_params.yaml"),
        epochs=args.epochs,
        device=device,  # dynamic device
        single_cls=args.single_cls,
        mosaic=args.mosaic,
        optimizer=args.optimizer,
        lr0=args.lr0,
        lrf=args.lrf,
        momentum=args.momentum
    )

    print(model)


"""

# Specify the path where you want to save the train.py file

train_file_path = '/content/HackByte_Dataset/train.py'

# Open the file in write mode and write the new script to it
with open(train_file_path, 'w') as f:
    f.write(train_script)

print(f"train.py file has been updated at {train_file_path}")
```
